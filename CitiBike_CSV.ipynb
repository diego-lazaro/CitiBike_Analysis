```python
# Dependencies and Setup
import pandas as pd

# Read Purchasing File and store into Pandas data frame
jan = pd.read_csv("./Resources/JC-201912-citibike-tripdata.csv")
feb = pd.read_csv("./Resources/JC-202001-citibike-tripdata.csv")
mar = pd.read_csv("./Resources/JC-202002-citibike-tripdata.csv")
apr = pd.read_csv("./Resources/JC-202003-citibike-tripdata.csv")
may = pd.read_csv("./Resources/JC-202004-citibike-tripdata.csv")
jun = pd.read_csv("./Resources/JC-202005-citibike-tripdata.csv")
jul = pd.read_csv("./Resources/JC-202006-citibike-tripdata.csv")
aug = pd.read_csv("./Resources/JC-202007-citibike-tripdata.csv")
sep = pd.read_csv("./Resources/JC-202008-citibike-tripdata.csv")
oct = pd.read_csv("./Resources/JC-202009-citibike-tripdata.csv")
nov = pd.read_csv("./Resources/JC-202010-citibike-tripdata.csv")
dec = pd.read_csv("./Resources/JC-202011-citibike-tripdata.csv")
```
